```python
import numpy as np
```

```python
def calculate_entropy(labels):
    # Count the occurrences of each class label
    _, counts = np.unique(labels, return_counts=True)
    
    # Calculate the probabilities for each class
    probabilities = counts / counts.sum()
    
    # Calculate entropy using the formula
    entropy = -np.sum(probabilities * np.log2(probabilities))
    
    return entropy

# Example dataset: a list of class labels
labels = [0, 1, 1, 0, 1, 1, 0, 0, 1, 0, 1, 1, 1 ,1, 1]
```

## Visualizing the relationship between information content, log(p), and probability, p

We can see that the information content decreases as the probability increases

```python
import numpy as np
import matplotlib.pyplot as plt

# Define probabilities (avoiding 0 to prevent log(0))
probabilities = np.linspace(0.001, 1, 1000)

# Calculate information content
information_content = -np.log2(probabilities)

# Plot
plt.figure(figsize=(8, 6))
plt.plot(probabilities, information_content, label='Information Content')
plt.xlabel('Probability of Event')
plt.ylabel('Information Content (bits)')
plt.title('Information Content vs. Probability')
plt.grid(True)
plt.legend()
plt.show()
```

## KL divergence

```python
import numpy as np

# Define the distributions
P = np.array([0.1, 0.6, 0.3])
Q = np.array([0.4, 0.4, 0.2])

# Compute the KL divergence
D_KL = np.sum(P * np.log(P / Q))

print(f"KL Divergence from Q to P: {D_KL:.4f}")
```

```python
# Compute the cross entropy
cross_entropy = -np.sum(P * np.log(Q))

print(f"Cross Entropy H(P, Q): {cross_entropy:.4f}")
```
